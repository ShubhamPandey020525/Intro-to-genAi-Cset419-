```python
import torch
from diffusers import StableDiffusionPipeline
import os

pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16
).to("cuda")
```

```python
cat_species = [
    "Abyssinian", "American Bobtail", "American Shorthair", "Balinese", "Bengal",
    "Birman", "Bombay", "British Shorthair", "Burmese", "Burmilla",
    "Chartreux", "Cornish Rex", "Devon Rex", "Egyptian Mau", "Exotic Shorthair",
    "Havana Brown", "Himalayan", "Japanese Bobtail", "Korat", "Lykoi",
    "Maine Coon", "Manx", "Munchkin", "Norwegian Forest Cat", "Ocicat",
    "Oriental", "Persian", "Ragamuffin", "Ragdoll", "Russian Blue",
    "Savannah", "Scottish Fold", "Selkirk Rex", "Siamese", "Siberian",
    "Singapura", "Somali", "Sphynx", "Tonkinese", "Toyger",
    "Turkish Angora", "Turkish Van"
]

for species in cat_species:
    species_folder = os.path.join(base_dir, species)
    os.makedirs(species_folder, exist_ok=True)

    prompt = f"Realistic photography of a {species} cat, high resolution, 4k, detailed fur"

    print(f"Generating for {species}...")
    for i in range(1, 11):
        image = pipe(prompt).images[0]
        # Label is the filename
        image.save(os.path.join(species_folder, f"{species}_{i}.png"))
```
